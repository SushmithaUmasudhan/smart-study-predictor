```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load Dataset from CSV

df = pd.read_csv(r"C:\Users\SUSHMITHA U\Downloads\archive (1)\score.csv")

print("Dataset Preview:")
print(df.head())

# Train Model

X = df[["Hours"]]
y = df["Scores"]

model = LinearRegression()
model.fit(X, y)

# Take User Input

hours = float(input("Enter study hours: "))

predicted_marks = model.predict([[hours]])[0]
predicted_marks = max(0, min(100, predicted_marks))
print(f"Predicted Marks: {predicted_marks:.2f}")
#Categorize Performance
if predicted_marks < 40:
    category = "Needs Improvement"
    suggestion = "Increase study time by at least 2 hours."
elif predicted_marks < 70:
    category = "Average"
    suggestion = "Try to revise more and practice regularly."
else:
    category = "Good"
    suggestion = "Great! Maintain your consistency."
# Output Results
print(f"\nPredicted Marks: {predicted_marks:.2f}")
print(f"Performance: {category}")
print(f"Suggestion: {suggestion}")
#Visualization
plt.scatter(df["Hours"], df["Scores"], label="Actual Data")
plt.plot(df["Hours"], model.predict(X), label="Regression Line")

plt.scatter(hours, predicted_marks, marker='x', label="Your Prediction")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks Predictor")
plt.legend()
plt.show()
```
